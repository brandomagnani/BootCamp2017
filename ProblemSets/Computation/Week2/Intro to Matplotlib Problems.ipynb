```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# Problem 1:

def get_var(n):
    
    rand_matrix = np.random.randn(n, n)
    mean_vec = np.mean(rand_matrix, axis=0)
    var = np.var(mean_vec)
    
    return var

def iterate():
    
    nn = np.int64(np.linspace(100, 1000, 10))
    var_list = []
    
    for i in nn:
        var_list.append(get_var(i))
    
    var_array = np.asarray(var_list)
    
    return var_array
```

```python
a = iterate()
print(a)
plt.plot(a)
plt.show()
```

```python
# Problem 2:

xx = np.linspace(-2*np.pi, 2*np.pi, 200)
fig, axes = plt.subplots(3, 1, figsize=(8, 12))
y = np.array([np.sin(xx), np.cos(xx), np.arctan(xx)])
titles = ["sin(x)", "cos(x)", "arctan(x)"]
for i in range(len(y)):
    
    axes[i].plot(xx, y[i], linewidth=2, alpha=0.6)
    axes[i].set_title(titles[i])
plt.show()
```

```python
# Problem 3:

xx1 = np.linspace(-2, 0.9, 200)
xx2 = np.linspace(1.1, 6, 200)

f = lambda x: 1 / (x-1)

fig = plt.figure(figsize=(6,6))
plt.plot(xx1, f(xx1), '--', linewidth=6, color='m')
plt.plot(xx2, f(xx2), '--', linewidth=6, color='m')
plt.ylim((-6, 6))
plt.show()
```

```python
# Problem 4:

xx = np.linspace(0, 2*np.pi, 200)
fig, ax = plt.subplots(figsize=(8,8))
plt.suptitle('PLOTS:', fontsize=18)

plt.subplot(221)
plt.plot(xx, np.sin(xx), color='g', linestyle='-')
plt.axis([0,2*np.pi, -2, 2])
plt.title('$sin(x)$')

plt.subplot(222)
plt.plot(xx, np.sin(2*xx), color='r', linestyle='--')
plt.axis([0,2*np.pi, -2, 2])
plt.title('$sin(2x)$')

plt.subplot(223)
plt.plot(xx, 2*np.sin(xx), color='b', linestyle='--')
plt.axis([0,2*np.pi, -2, 2])
plt.title('$2sin(x)$')

plt.subplot(224)
plt.plot(xx, 2*np.sin(2*xx), color='m', linestyle=':')
plt.axis([0,2*np.pi, -2, 2])
plt.title('$2sin(2x)$')
plt.show()
```

```python
# Problem 5:

x = np.linspace(-2*np.pi, 2*np.pi, 200)
y = x.copy()
X, Y = np.meshgrid(x, y)
Z = (np.sin(X) * np.sin(Y)) / (X * Y)

fig, ax = plt.subplots(figsize=(16,8))

plt.subplot(121)
plt.pcolormesh(X, Y, Z, cmap="viridis" )
plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
plt.colorbar()

plt.subplot(122)
plt.contour(X, Y, Z, 400, cmap="magma" )
plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
plt.colorbar()
```
