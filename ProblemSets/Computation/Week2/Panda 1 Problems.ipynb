```python
# Problem 2:

N = 100
s = np.zeros((N, 5))
NN = [100, 1000, 10000]


for i in range(5):
    s[1:,i] = np.random.binomial(1, 0.5, size=(N-1,))*2-1
s = np.matrix.cumsum(s, axis=0)
s = pd.DataFrame(s, index=range(N), columns=["RW1", "RW2", "RW3", "RW4", "RW5"])

fig, ax = plt.subplots(figsize=(16,12))

plt.subplot(2,2,1)
s.RW1.plot()
s.RW2.plot()
s.RW3.plot()
s.RW4.plot()
s.RW5.plot()
plt.show()

for t, n in enumerate(NN):
    s_bias = np.zeros(n)
    s_bias = np.random.binomial(1, 0.505, size=(n-1,))*2-1
    s_bias = np.cumsum(s_bias)
    s_bias = pd.Series(s_bias)
    plt.subplot(2,2,t+2)
    s_bias.plot()
    plt.show()
```
