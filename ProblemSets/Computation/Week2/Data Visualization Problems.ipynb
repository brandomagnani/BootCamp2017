```python
import numpy as np
from scipy.special import binom
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
# Problem 1:

data = np.load("/Users/ildebrandomagnani/Desktop/anscombe.npy")

x1 = data[:,0]
x2 = data[:,2]
x3 = data[:,4]
x4 = data[:,6]
y1 = data[:,1]
y2 = data[:,3]
y3 = data[:,5]
y4 = data[:,7]

f = lambda x: 0.5*x +3
xx = np.linspace(0, 20, 100)

fig, ax = plt.subplots(figsize=(16,16))

plt.subplot(221)
plt.plot(x1, y1, "o")
plt.plot(xx, f(xx))

plt.subplot(222)
plt.plot(x2, y2, "o")
plt.plot(xx, f(xx))

plt.subplot(223)
plt.plot(x3, y3, "o")
plt.plot(xx, f(xx))

plt.subplot(224)
plt.plot(x4, y4, "o")
plt.plot(xx, f(xx))
```

```python
# Problem 2:

bnv = lambda x, n, v: binom(n,v) * (x**v) * ((1-x)**(n-v))
xx = np.linspace(0, 1, 100)
N = 3
i = 1

fig, ax = plt.subplots(figsize=(16,22))


for n in range(N+1):
    for v in range(n+1):
        plt.subplot(5,2,i)
        plt.plot(xx, bnv(xx, n, v))
        plt.ylim(0,1.2)
        plt.title("Bernstein("+str(n)+ ", " +str(v)+")")
        i+=1
```

```python
# Problem 3:

data = np.load("/Users/ildebrandomagnani/Desktop/MLB.npy")
height = data[:,0]
weight = data[:,1]
age    = data[:,2]

fig, ax = plt.subplots(figsize=(12,10))

plt.subplot(111)
plt.scatter(weight, height, c=age)
clb = plt.colorbar()
clb.set_label("Age (years)")
plt.xlabel("Weight (pounds)")
plt.ylabel("Height (inches)")
```

```python
# Problem 4:

data = np.load("/Users/ildebrandomagnani/Desktop/earthquakes.npy")

year      = data[:,0]
magnitude = data[:,1]
longitude = data[:,2]
latitude  = data[:,3]

fig, ax = plt.subplots(figsize=(18,8))

plt.subplot(221)
plt.plot(year, magnitude, ".")
plt.xlabel("Year")
plt.ylabel("Magnitude (Richter Scale)")

plt.subplot(222)
plt.hist(year, bins=11,histtype="stepfilled")
plt.xlabel("Year")
plt.ylabel("Number of earthquakes")

plt.subplot(223)
plt.hist(magnitude, bins=10)
plt.xlabel("Magnitude (Richter)")
plt.ylabel("Number of earthquakes")
plt.xlim(4.9, 8)

plt.subplot(224)
plt.scatter(longitude, latitude, c=magnitude)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
clb = plt.colorbar()
clb.set_label("Magnitude (Richter)")
plt.xlim(4.9, 8)
```

```python
# Problem 5:

x = np.linspace(-4, 4, 200)
y = x.copy()
X, Y = np.meshgrid(x, y)

Z = (1-X)**2 + 100*(Y-X**2)**2

fig, ax = plt.subplots(figsize=(16, 8))

plt.subplot(121)
plt.pcolormesh(X, Y, Z, cmap="plasma", norm=mpl.colors.LogNorm())
plt.colorbar()

plt.subplot(122)
plt.contourf(X, Y, Z, 6, cmap="plasma", norm=mpl.colors.LogNorm())
plt.colorbar()
```
